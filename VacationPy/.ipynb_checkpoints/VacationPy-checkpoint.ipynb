### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Find the first hotel for each city located within 5000 meters of your coordinates

h = []

for x in range(len(iw1["City"])):

    lat = iw1.loc[x]["Lat"]
    lng = iw1.loc[x]["Lng"]

    p = {
        "location": f"{lat},{lng}", 
        "types": "lodging",
        "radius": 5000,
        "key": g_key
    }

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"   

    h_req = requests.get(base_url, params=p)
    h_resp = h_req.json()

    try:
        h.append(h_resp["results"][0]["name"])
    except:
        h.append("Hotel located within 5,000 meters of coord. is not available.")

# Dataframe with nearest hotel
iw1["Hotel Name"] = h
iw1
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in iw1.iterrows()]
locations = iw1[["Lat", "Lng"]]
```

```python
# Plot the hotels on top of the humidity heatmap

mk = gmaps.marker_layer(locations, info_box_content=hotel_info)

hm3.add_layer(mk)

hm3
```
