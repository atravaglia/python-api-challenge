```python
### IMPORTANT - API KEY FILES NOT INCLUDED FOR SAFETY REASONS.


# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "../output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Perform weather check from API 
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{url}appid={weather_api_key}&units={units}&q="

api0 = requests.get(query_url + city)
api1 = api0.json()
print(json.dumps(api1, indent=2))
```

```python
print(requests.get(query_url + city))
```

```python
# Establish the list of selected city parameters 
city_name = []
lat = []
lng = []
temp_max = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []

x = 0

# Perform loop and increment by one at each loop

print(f"X-X-- City Log --X-X")

for city in cities:
    
    try:
        c1 = requests.get(query_url + city).json()
        city_name.append(c1["name"])
        country.append(c1["sys"]["country"])
        date.append(c1["dt"])
        lat.append(c1["coord"]["lat"])
        lng.append(c1["coord"]["lon"])
        temp_max.append(c1["main"]["temp_max"])
        humidity.append(c1["main"]["humidity"])
        cloudiness.append(c1["clouds"]["all"])
        wind_speed.append(c1["wind"]["speed"])
        city_record = c1["name"]
        print(f"List of City Records {x} | {city_record}")
        
        x += 1
        
    except:
        print("disregard unavailable data") 
    
    continue


```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create dictionary
ow = {
    "City": city_name,
    "Lat": lat,
    "Lng": lng, 
    "Temp Max": temp_max,
    "Humidity": humidity,
    "Cloudiness": cloudiness,
    "Wind Speed": wind_speed,
    "Country": country,
    "Date": date
}

ow0 = pd.DataFrame(ow)
ow0.count()
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames
northern = ow0.loc[pd.to_numeric(ow0["Lat"]).astype(float) > 0, :]
southern = ow0.loc[pd.to_numeric(ow0["Lat"]).astype(float) < 0, :]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Calculate linear regression
x_values = pd.to_numeric(northern['Lat']).astype(float)
y_values = pd.to_numeric(northern['Temp Max']).astype(float)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(f"Equation for linear regression: {line_eq}")

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Lat')
plt.ylabel('Max Temperature')
plt.title('Northern Hemisphere - Max Temp vs. Latitude Linear Regression')

print(f"Coefficient for r-squared: {rvalue}")

print("Northern Hemisphere: The plot shows that the maximum temperature decreases as the latitude increases")
print ("(from equator to north pole). This is reflected by a negative correlation.")

plt.show()
```
